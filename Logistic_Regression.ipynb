```python
hz = np.linspace(0.01, 1, 200)
```

If true class is 1(y(i) = 1)

```python
L1 = -np.log(hz)
```

if true class is 0(y(i) = 0)

```python
L2 = -np.log(1-hz)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

axs[0].plot(hz, L1, color='blue')
axs[0].set_title('If true class is 1')
axs[0].set_xlabel('h(z)')
axs[0].set_ylabel('Loss(h(z))')
axs[0].grid(alpha=0.3)

axs[1].plot(hz, L2 , color= 'red')
axs[1].set_title('If true class is 0')
axs[1].set_xlabel('h(z)')
axs[1].set_ylabel('Loss(h(z))')
axs[1].grid(alpha=0.3)

```
